# we are goin to use the Hacker posts database to extract insight and answer the following two question
## A challenge in this analysis is to not use numpy or pandas and only rely on lists and dict

1) Do Ask HN or Show HN receive more comments on average?

2) Do posts created at a certain time receive more comments on average?

```python
from csv import reader
import datetime as dt
```

```python
with open('hacker_news.csv') as open_file:
    hn= list(reader(open_file))
hn[:5]
```

```python
#Remove the header from the dataset
header = hn[0]
hn=hn[1:]
```

```python
ask_posts=[]
show_posts =[]
other_posts=[]
for row in hn:
    title=row[1]
    if title.lower().startswith('ask hn'):
        ask_posts.append(row)
    elif title.lower().startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
print('{} ask posts, {} show posts, {} other posts'.format(len(ask_posts),len(show_posts),len(other_posts)))
```

```python
## find avg number of comments on ask posts
total_ask_comments=0
for post in ask_posts:
    comment=int(post[4])
    total_ask_comments+=comment
avg_ask_comments = round(total_ask_comments/len(ask_posts))
print('Avg number of comments on ASK posts is ' , avg_ask_comments)

## find avg number of comments on show posts
total_show_comments=0
for post in show_posts:
    comment=int(post[4])
    total_show_comments+=comment
avg_show_comments = round(total_show_comments/len(show_posts))
print('Avg number of comments on SHOW posts is ' , avg_show_comments)
```

```python
# Calculate the number of comments per hour and the nmber of posts per hour
# Calculate the average number of comments per post per hour
counts_by_hour ={}
comments_by_hour={}
temp = '%m/%d/%Y %H:%M'
for post in ask_posts:
    date_time= post[6]
    count = int(post[4])
    date_time = dt.datetime.strptime(date_time,temp)
    hour = date_time.hour
    if hour not in comments_by_hour:
        comments_by_hour[hour]= count
        counts_by_hour[hour] = 1
        hour = date_time.hour
    elif hour in comments_by_hour:
        comments_by_hour[hour]+= count
        counts_by_hour[hour] += 1
avg_com_per_post_per_hr=[]
for hr in comments_by_hour:
    avg_com_per_post_per_hr.append((hr,round(comments_by_hour[hr]/counts_by_hour[hr])))
avg_com_per_post_per_hr.sort(reverse=True,key = lambda x: x[1])
print('(hr, avg)')
for i in avg_com_per_post_per_hr:
    print(i)
    
```

## It seems that posts soliciting advice get more attention, Also posts around 2am or 3pm get the most attention this makes sense because not everyone is in the same time zone.
